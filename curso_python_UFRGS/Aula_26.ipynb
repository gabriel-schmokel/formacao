## Limpeza de Dados

1. células vazias: o que fazer com elas?
2. dados corretos, mas em formato errado: o que fazer?
3. dados errados: o que fazer?
4. dados duplicados o que fazer?


### Células vazias

```python
# Remover todas as linhas que contenham as células vazias.
# Para remover todas as linhas sem elementos utiliza-se o atributo .dropna(...)

df1 = df.dropna()
print(df.info())
print()
print(df1.info())
print()

# A remoção dessas linhas pode ser feita no máximo até 10% total das linhas dos dados
# Para vereficar o percentual deve-se fazer uma regra de três.
print('Percentual: ')
print(1 - len(df1)/len(df))
```

```python
# Substituir células vazias por valores específicos
df1 = df.copy()

# tomando a média da coluna calorias ou a mediana
xmed = df1['Calories'].mean()
xmedian = df1['Calories'].median()
print('Média e mediana: ')
print(xmed, xmedian)
print()
# substituindo a média nos elementos NaN através do atributo .fillna(...,...)
# primeiro argumento é o número a ser substituido.
# segundo argumento é para validar a substituição.
df1['Calories'].fillna(xmedian,inplace=True)

print(df1.info())
print()

# Substituir célula vazia na coluna 'Date' pela data faltante.
df1.loc[25,'Date'] = '2021-01-25'
print(df1.info())
print()

# Removendo linhas com células vazias na coluna 'Date'
df1.dropna(subset=['Date'], inplace = True)
print(df1.info())
```

### Dados corretos em formato (tipo) errado


```python
# o atributo .to_datetime(...) permite converter os ditpos daddos no formato datetime

df1['Date'] = pd.to_datetime(df1['Date'])
print(df1.head(25))
```

### Dados errados (valores incorretos)

```python
# Em uma interação podemos remover as linhas que julgamos incorretas através do uso de um filtro
# Para remover a linha usamos o atributo .drop(...,....)
# Primeiro argumento é a linha.
# Segundo argumento é pra validar a exclusão da linha.

MAX_DURATION = 120

for i in df1.index:
  if df1.loc[i,'Duration'] > MAX_DURATION:
    df1.drop(i, inplace = True)

print(df1.to_string())
```

### Dados Duplicados

```python
# Para visualizar se os dados são duplicadoso utilizamos o atributo .duplicated()
# Irá retornar uma lista de True or False

print(df1.duplicated())
print()

print(df1.loc[df1.duplicated()])
print()

# Para remover os dados duplicados usamos
df1.drop_duplicates(inplace=True)
print(df1.to_string())
```
