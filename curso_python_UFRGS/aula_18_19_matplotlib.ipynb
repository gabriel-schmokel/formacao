# Matplotlib

Módulo: **matplotlib**

Matplotlib é uma biblioteca de software para criação de gráficos e visualizações de dados em geral, feita para e da linguagem de programação Python e sua extensão de matemática NumPy. 

Originalmente criada pelo biólogo e neurocientista americano John D. Hunter.

Oferece uma interface de programação orientada a objetos para incluir gráficos em aplicações usando **toolkits** de interface gráfica, como **Tkinter**, **WxPython**, **Qt** ou **GTK**.



## Gráficos xy

### Função: **plt.plot(x,y)**

```python
# Gerando dados 
x = [10,20,30,40]
y = [100,400,900,1600]

# Plotando:
# Para plotar gráficos devemos importar o método pyplot contido na biblioteca matplotlib
import matplotlib.pyplot as plt

# A função .plot() produz o gráfico.
# Recebe dois argumentos: primeiro corresponde aos dados do eixo 'x', o segundo aos dados do eixo 'y'.
# A função .plot() retorna um objeto lista que é salvo na memória do computador, podemos essa infformação a uma variável. 
# Não é possível printar essa informação.
o = plt.plot(x,y)
print(type(o))

# para exibir o gráfico, da lista armazenada na memória, ou da variável que contém essa lista, usamos a função .show().
# o argumento fa função .show() é a variável, a qual foi atribuida a lista.
# por default, ou seja, sem argumentos, será exibido a lista contida na memória gerada pela função .plot().

plt.title('Título')
plt.xlabel('x')
plt.ylabel('y')
plt.show(o)
```

### Múltiplas Curvas num Mesmo Gráfico

```python
import numpy as np

# Gerando dados 
# Agora produzimos um array e não uma lista. 
x = np.arange(-np.pi,np.pi, 0.1)
y1 = x
y2 = np.sin(x)
y3 = np.cos(x)

print(x)

# Plotando múltiplas curvas num mesmo gráfico.
# Por default as cores são geradas diferestes entre as curvas.


# Modo 01: com mais linhas de código
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)

plt.title('Gráfico 2a')
plt.xlabel('x')
plt.ylabel('y')

plt.show()

# Modo 02: com menos linhas de código
plt.plot(x,y1, x, y2, x, y3)
plt.title('Gráfico 2a')
plt.xlabel('x')
plt.ylabel('y')

plt.show()

```

### Alterando a Espessura das Linhas

```python
import numpy as np

# Gerando dados 
# Dados gerados por array. 
x = np.arange(-np.pi,np.pi, 0.1)
y1 = x
y2 = np.sin(x)
y3 = np.cos(x)

print(x)

# Alterando a espessura das múltiplas curvas num mesmo gráfico.
# O argumento lw = ... na função .pĺot() é o que define a espessura das linhas.
# lw significa line width
# Por default as linhas das curvas tem o tamanho 1.


# Modo 01: com mais linhas de código
plt.plot(x,y1,lw=1)
plt.plot(x,y2,lw=2)
plt.plot(x,y3,lw=3)

plt.title('Gráfico 2a')
plt.xlabel('x')
plt.ylabel('y')

plt.show()

# Modo 02: com menos linhas de código
# O modo 02 não permite costumizar cada curva individualmente, ele define para todas as linhas/curvas.
plt.plot(x,y1, x, y2, x, y3, lw =4)
plt.title('Gráfico 2a')
plt.xlabel('x')
plt.ylabel('y')

plt.show()

```

## Diagramas de Dispersão

Os diagramas de dispersão ou gráficos de dispersão são representações de dados de duas (tipicamente) ou mais variáveis que são organizadas em um gráfico. O gráfico de dispersão utiliza coordenadas cartesianas para exibir valores de um conjunto de dados. Os dados são exibidos como uma coleção de pontos, cada um com o valor de uma variável determinando a posição no eixo horizontal e o valor da outra variável determinando a posição no eixo vertical (em caso de duas variáveis).

Descrito pela primeira vez por Francis Galton, o gráfico de dispersão é usado para verificar se existe relação de causa e efeito entre duas variáveis de natureza quantitativa (variáveis que podem ser medidas ou contadas). Isto não prova que uma variável afeta outra variável, mas determina se existe relação e qual a intensidade da relação entre elas. Por exemplo, o gráfico de dispersão possibilita construir uma regressão linear, determinando—se uma reta que aponta a relação entre duas variáveis e indica a função que dá o comportamento da relação entre elas.


```python
import numpy as np
import matplotlib.pyplot as plt

# Exemplo 1) 
# Gerar uma distribuição aleatória de 'n' pontos, 
# distribuidas dentro de uma área cirular, com centro (0,0) e raio igual a r = 1.

# Gerar dados
n = 1000
x = 2*np.random.rand(n)  -1  # [0,1]
y = 2*np.random.rand(n)  -1 
filtro = np.sqrt(x**2 +y**2) <= 1

# filtro é composto por um array TRUE ou FALSE
x = x[filtro]          # retorna o array com os valores de 'x' válidos                 
y = y[filtro]          # retorna o array com os valores de 'y' válidos

# visualização
# A função scatter() é a específica para os diagramas de dispersão, apesar de podermos usar plt.plot(x,y,'.') 
# Função scattter recebe como arugmento dois arrays 1D que contém a posição dos pontos.
# Primeiro array é a posição na abscissa, e o segundo informa a posição no eixo ordenado.
# devido a grande quantidade de pontos 'n', deve-se diminuir o tamanho dos pontos.
plt.scatter(x,y)
plt.title('Diagrama de Dispersão')
plt.xlabel('x')
plt.ylabel('y')

# outro argumento que .scatter() recepe é a cor dos pontos, dado por c = '...' ou color
# o marcador das bordas dos pontos é dado por marker 
plt.scatter(x,y,c='g', marker = '.')

# a função plt.gca() retorna um objeto contendo as informação dos eixos do gráfico.
# podemos atribuir essas informações a uma variável.
ax = plt.gca()       # chamamos a variável de ax para relacionar com eixos.


# podemos ajustar o aspecto = < comprimento do eixo da escala 'y'> / < comprimento do eixo da escala 'x'>
# entendemos como aspecto a largura dos eixos x e y.
# mudança da razão é dada pelo atributo set_aspect(1)
# se dejamos que as duas escalas tenham o mesmo comprimento devemos colocar como argumento 1 no atributo.
ax.set_aspect(1)




plt.show()



```

## Diagramas de Barras ou Gráfico de Barras

Usado para representar variáveis categóricas.

O gráfico de barras é um gráfico com barras retangulares e comprimento proporcional aos valores que ele apresenta. As barras podem ser desenhadas na vertical ou na horizontal.

Este tipo de representação utiliza barras tanto verticais quanto horizontais para ilustrar comparações (sendo este primeiro mais conhecido como gráfico de colunas). Um eixo do gráfico mostra especificamente o que está sendo comparado enquanto o outro eixo representa valores discretos. Alguns gráficos de barra apresentam barras agrupadas em grupos (gráficos de barras agrupadas) e outros mostram as barras divididas em sub-partes para mostrar efeito acumulativo (gráficos de barras empilhadas). 

```python
import matplotlib.pyplot as plt


# Gerar dados:  
Tipos = ['Tipo 1', 'Tipo 2', 'Tipo 3', 'Tipo 4', 'Tipo 5']
Valores = [10,15,27,13,5]


# Modo 01: não é utilizado usando a função .plot() 
# Não gera uma simplificação da visualização da informação simples usando para as variáveis categóricas.    
plt.plot(Tipos, Valores)
plt.title('.func()')
plt.xlabel('Tipos')
plt.ylabel('Valores')
plt.show()

print()

# Modo 02: não é utilizado usando a função .scatter()
# Não gera uma simplificação da visualização da informação simples usando para as variáveis categóricas.    
plt.scatter(Tipos, Valores)
plt.title('.scatter()')
plt.ylabel('Valores')
plt.show()

print()

# Modo 03 - a: Gráfico de Barras
# usando a função .bar()
# Gera uma simplificação da visualização da informação para as variáveis categóricas.    
plt.bar(Tipos, Valores)
plt.title('Diagrama de Barras 01 - .func()')
plt.ylabel('Valores')
plt.show()

print()

# Modo 03 - b: Gráfico de Barras
# usando a função .bar()
# Gera uma simplificação da visualização da informação para as variáveis categóricas.    
# Para alterar as cores das barras colocamos um terceiro argumento, contendo uma lista com as cores que desejamos que cada barra tenha.
# Para alinhar a barra na borda das variáveis categóricas, colocamos o argumento align = 'edge'. Existe outras formas de alinhas, deve-se conferir
# O argumento edgecolor permite alterar a cor da borda das barras.

plt.bar(Tipos,
        Valores,
        color = ['blue', 'green', 'red', 'yellow', 'cyan'], 
        #align = 'edge',
        edgecolor = 'black'
        )
plt.title('Diagrama de Barras 02 - .func()')
plt.ylabel('Valores')
plt.show()
```

## Gráficos de Pizza

ou **gráficos de setores** ou **gráfico circular**

```python
import matplotlib.pyplot as plt


# Gerar dados:  
Tipos = ['Tipo 1', 'Tipo 2', 'Tipo 3', 'Tipo 4', 'Tipo 5']
Valores = [10,15,27,13,5]

# Gráfico de pizza - chamado através da função pie()
# O argumento 'Valores' corresponde a forma como vamos dividir a pizza - passa-se uma lista
# para rotular os valores das fatias da lista, passamos uma lista a 'labels'
# para exibir uma sombra na borda da pizza colocamos como 'True' o argumento shadow
# podemos colocar o percentual dos valores com argumento 'autopct'. O valor do 'autopct' é escrito da mesma foma para exibir o formato das variáveis em 'c'
# o argumento startangle é o ângulo de início pelo qual será dada a sequência das fatias contendo os dados com seus tipos.
# podemos afastar as fatias passando como argumento explode, onde o valor desse argumento é uma lista.

# os valores da lista correspondem ao módulo que cada um dos 'Tipos' ficaram separados do centro da pizza.
explosao = [0,0,0.1,0,0]

plt.pie(Valores, 
        labels = Tipos, 
        shadow = True,
        startangle = 90,
        explode = explosao, 
        autopct = '%4.1f')
plt.title('Diagrama de pizza - .pie()')
plt.show()

print()
```

## Histogramas

O histograma, também conhecido como distribuição de frequências, é a representação gráfica em colunas ou em barras (retângulos) de um conjunto de dados previamente tabulado e dividido em classes uniformes ou não uniformes. A base de cada retângulo representa uma classe. A altura de cada retângulo representa a quantidade ou a frequência absoluta com que o valor da classe ocorre no conjunto de dados para classes uniformes ou a densidade de frequência para classes não uniformes. Importante ferramenta da estatística, o histograma também é uma das chamadas sete ferramentas da qualidade.

```python
import numpy as np

# gerando dados:
distr = np.random.randn(1000)         # Gerando uma distribuição normal de pontos


# EXEMPLO 01
# Utilizado a função .hist() para gerar o histograma
# Utilizado para processos contínuos, caso contrário usar .bar() 
# Recebe como argumento uma lista e irá separar cada valor em intervalos, para então contabilizar um dado. No exemplo abaixo estamos usando a lista como sendo 'distr'

plt.hist(distr, density = 'True')     
plt.title('Histograma 01')
plt.xlabel('x')
plt.ylabel('Número de Ocorrências')                  
plt.show()


# EXEMPLO 02
# Utilizado a função .hist() para gerar o histograma
# Utilizado para processos contínuos, caso contrário usar .bar() 
# Recebe como argumento uma lista e irá separar cada valor em intervalos, para então contabilizar um dado. No exemplo abaixo estamos usando a lista como sendo 'distr'
# 'bins' delimita o número de barras que iremos ter no histograma. É indicado gerar um numero de barras respeitando a seguinte expressão: x tem que ser menor sqrt(n), onde n é o número de dados
# O argumento 'density' define se queremos representar a normalização dos dados, ou seja, trabalharmos com a porcentagem deles.

plt.hist(distr, bins = 50, density = 'True')     
plt.title('Histograma 02')
plt.xlabel('x')
plt.ylabel('Número de Ocorrências')                  
plt.show()



# EXEMPLO 03
# Utilizado a função .hist() para gerar o histograma
# Utilizado para processos contínuos, caso contrário usar .bar() 
# Recebe como argumento uma lista e irá separar cada valor em intervalos, para então contabilizar um dado. No exemplo abaixo estamos usando a lista como sendo 'distr'
# 'bins' delimita o número de barras que iremos ter no histograma. É indicado gerar um numero de barras respeitando a seguinte expressão: x tem que ser menor sqrt(n), onde n é o número de dados
# O argumento 'color' modifica a cor das barras.
# O argumento 'density' define se queremos representar a normalização dos dados, ou seja, trabalharmos com a porcentagem deles.

plt.hist(distr, bins = 50, color = 'yellow', density = 'True')     
plt.title('Histograma 03')
plt.xlabel('x')
plt.ylabel('Número de Ocorrências')                  
plt.show()


# EXEMPLO 04
# Utilizado a função .hist() para gerar o histograma
# Utilizado para processos contínuos 
# Recebe como argumento uma lista e irá separa cada valor em intervalor para então contabilizar um dado. No exemplo abaixo estamos usando a lista como sendo 'distr'
# 'bins' delimita o número de barras que iremos ter no histograma. É indicado gerar um numero de barras respeitando a seguinte expressão: x tem que ser menor sqrt(n), onde n é o número de dados
# O argumento 'color' modifica a cor das barras.
# O argumento 'density' define se queremos representar a normalização dos dados, ou seja, trabalharmos com a porcentagem deles.
# Podemos mudar o contorno/tipo do histograma com o argumento histtype = ...

plt.hist(distr, bins = 50, color = 'red', density = 'True', histtype = 'step')     
plt.title('Histograma 04')
plt.xlabel('x')
plt.ylabel('Número de Ocorrências')                  
plt.show()
```

```python
# ARRANJO 2X2
# FUNÇÃO .subplot(...) faz plot de múltiplos gráficos 
# como argumento a subplot passamos primeiro o arranjo e posteriormente a numeração do ordenamento das figuras.

import numpy as np

# gerando dados:
distr = np.random.randn(1000)         # Gerando uma distribuição normal de pontos


# EXEMPLO 01
# Utilizado a função .hist() para gerar o histograma
# Utilizado para processos contínuos, caso contrário usar .bar() 
# Recebe como argumento uma lista e irá separar cada valor em intervalos, para então contabilizar um dado. No exemplo abaixo estamos usando a lista como sendo 'distr'
plt.subplot(221)
plt.hist(distr, density = 'True')     
plt.title('Histograma 01')
plt.xlabel('x')
plt.ylabel('Número de Ocorrências')                  

# EXEMPLO 02
# Utilizado a função .hist() para gerar o histograma
# Utilizado para processos contínuos, caso contrário usar .bar() 
# Recebe como argumento uma lista e irá separar cada valor em intervalos, para então contabilizar um dado. No exemplo abaixo estamos usando a lista como sendo 'distr'
# 'bins' delimita o número de barras que iremos ter no histograma. É indicado gerar um numero de barras respeitando a seguinte expressão: x tem que ser menor sqrt(n), onde n é o número de dados
# O argumento 'density' define se queremos representar a normalização dos dados, ou seja, trabalharmos com a porcentagem deles.
plt.subplot(222)              
plt.hist(distr, bins = 50, density = 'True')     
plt.title('Histograma 02')
plt.xlabel('x')
plt.ylabel('Número de Ocorrências')    

# EXEMPLO 03
# Utilizado a função .hist() para gerar o histograma
# Utilizado para processos contínuos, caso contrário usar .bar() 
# Recebe como argumento uma lista e irá separar cada valor em intervalos, para então contabilizar um dado. No exemplo abaixo estamos usando a lista como sendo 'distr'
# 'bins' delimita o número de barras que iremos ter no histograma. É indicado gerar um numero de barras respeitando a seguinte expressão: x tem que ser menor sqrt(n), onde n é o número de dados
# O argumento 'color' modifica a cor das barras.
# O argumento 'density' define se queremos representar a normalização dos dados, ou seja, trabalharmos com a porcentagem deles.

plt.subplot(223)               
plt.hist(distr, bins = 50, color = 'yellow', density = 'True')     
plt.title('Histograma 03')
plt.xlabel('x')
plt.ylabel('Número de Ocorrências')   

# EXEMPLO 04
# Utilizado a função .hist() para gerar o histograma
# Utilizado para processos contínuos 
# Recebe como argumento uma lista e irá separa cada valor em intervalor para então contabilizar um dado. No exemplo abaixo estamos usando a lista como sendo 'distr'
# 'bins' delimita o número de barras que iremos ter no histograma. É indicado gerar um numero de barras respeitando a seguinte expressão: x tem que ser menor sqrt(n), onde n é o número de dados
# O argumento 'color' modifica a cor das barras.
# O argumento 'density' define se queremos representar a normalização dos dados, ou seja, trabalharmos com a porcentagem deles.
# Podemos mudar o contorno/tipo do histograma com o argumento histtype = ...
plt.subplot(224)                
plt.hist(distr, bins = 50, color = 'red', density = 'True', histtype = 'step')     
plt.title('Histograma 04')
plt.xlabel('x')
plt.ylabel('Número de Ocorrências')  


# AJUSTANDO OS SUB PLOTS NA FIGURA
# função sub_plots_adjust(...) permite ajustar as margens a esquerda, a direita, acima, abaixo, a separação horizontal e vertical.
plt.subplots_adjust(top = 0.92,
                     bottom = 0.08,
                     left = 0.10,
                     hspace = 0.55,
                     wspace = 0.5)

plt.show()
```

## Textos Em Gráficos 

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerando dados:
mu = 3.5       # média da distribuição
sigma = 1.7    # variação da distribuição
distr = sigma*np.random.randn(10000) + mu

# visualização
plt.hist(distr, bins=50, color = 'cyan', density = True)

# O argumento fontsize em .title(...), .xlabel(...) e .ylabel(...) pertmite alterar o tamanho da fonte
# O argumento color permite alterar a cor da fonte
plt.title('Distribuição Normal ($\mu, \sigma$)', fontsize = 20, color = 'green')
plt.xlabel('x', fontsize = 16, color = 'blue')
plt.ylabel('Densidade', fontsize = 16, color = 'red')


# .text(...) permite escrever um texto dentro do gráfico
# o primeiro e o segundo argumento informa a posição horizontal e vertical onde o texto deve ficar
# o terceiro argumento é o texto que será exibido no gráfico.
plt.text(-2.5,0.20, 'Distribuição Normal')
plt.text(-2.5,0.17, '$\mu=3.5$, $\sigma$ = 1.7')

plt.grid(True)
plt.show
```

```python
# Anotações em Gráficos

import numpy as np

# Gerando dados:
A = 1.              # amplitude
f = 0.01            # frequência (Hz)
phi = 0.            # fase

t = np.arange(0,1000,1)              # tempo em segundos
y = A*np.sin(2*np.pi*f*t +phi)      # evolução temporal de uma oscilação não amortecida

Amort = y*np.exp(-0.005*t) 

# Gráfico 01: oscilação não amortecida
plt.subplot(211)
plt.plot(t,y)
plt.title('Oscilação não amortecida')
plt.xlabel('tempo (s)')
plt.ylabel('Amplitude (m)')

# Gráfico 02: oscilação amortecida
plt.subplot(212)
plt.plot(t,Amort)
#curva do envelope do amortecimento
plt.plot(t,np.exp(-0.005*t), color = 'r')
plt.plot(t,-np.exp(-0.005*t), color = 'r')
# legendas
plt.title('Oscilação amortecida')
plt.xlabel('tempo (s)')
plt.ylabel('Amplitude (m)')

# CÓDIGO PARA FAZER UMA ANOTAÇÃO NO GRÁFICO
# deve-se usar a função .annotate()
# argumentos 'xy' é um ponto do gráfico de referência aos outros parâmetros.
# argumento 'xytext' altera a posição do texto onde irá ficar nas coordenadas x e y.
# o argumento arrowprops permite fazer uma flecha, a ele é atribuido um dicionário que contém a cor e comprimento da flecha. 
# existem outras proprietodades nesse argumento 'arrowprops' que permitem alterar a espessura da flecha. 

plt.annotate('Amortecimento',
             xy = (300, 0.2),
             xytext = (650 , 0.7),
             arrowprops = dict(facecolor = 'black', shrink = 0.07)
             )
      

# ajuste das multiplas figuras 
plt.subplots_adjust(top = 0.92,
                     bottom = 0.08,
                     left = 0.10,
                     hspace = 0.75,
                     wspace = 0.5)

plt.show()

```

## Gráficos Logarítimicos

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerando dados
# Para função normal com os respectivos argumentos:
# 'loc' corresponde a média
# size é o número de pontos aleatórios gerados
# 'scale' é equivalente ao sigma.    
y = np.random.normal(loc = 0.5, scale = 0.4, size = 4000)
y = y[(y>0) & (y<1)]
x = np.arange(len(y))
y.sort()

# função que cria a figura
plt.figure()

#  GRÁFICO 01 - linear-linear
plt.subplot(221)
plt.plot(x,y)
plt.xlabel('x (linear)')
plt.ylabel('y (linear)')


#  GRÁFICO 02 - linear-log
plt.subplot(222)
plt.plot(x,y)
plt.yscale('log')             # transforma o eixo y em uma escala logaritimica  
plt.xlabel('x (linear)')
plt.ylabel('y (log)')

#  GRÁFICO 03 - log-log
plt.subplot(223)
plt.plot(x,y)
plt.yscale('log')             # essa função transforma o eixo y em uma escala logaritimica
plt.xscale('log')             # transforma o eixo x em uma escala logaritimica  
plt.xlabel('x (log)')
plt.ylabel('y (log)')

#  GRÁFICO 04 - log-linear
plt.subplot(224)
plt.plot(x,y)
plt.xscale('log')             # transforma o eixo x em uma escala logaritimica
plt.yscale('log')             # transforma o eixo y em uma escala logaritimica  
plt.xlabel('x (log)')
plt.ylabel('y (linear)')

# ajuste das multiplas figuras 
plt.subplots_adjust(top = 0.92,
                     bottom = 0.08,
                     left = 0.10,
                     hspace = 0.75,
                     wspace = 0.5)

plt.show()

```
